```python
# Removing the "Not Provided" values and  Changing the data type of salaries from string to float
df.dtypes
```

```python
col=['BasePay','OvertimePay','OtherPay','Benefits','EmployeeName']
for a in col:
    df[a].replace({'Not Provided':np.nan,'Not provided':np.nan},inplace=True)
df
```

```python
col=['BasePay','OvertimePay','OtherPay','Benefits']
for a in col:
    df[a]=df[a].astype(float)
df.dtypes
```

```python
col=['BasePay','OvertimePay','OtherPay','Benefits']
for a in col:
    print(a)
    print('Minimum {} = {}'.format(a,df[a].min()))
    print('Maximum {} = {}'.format(a,df[a].max()))
    print('Average {} = {}'.format(a,df[a].mean()))
    print('\n')
```
